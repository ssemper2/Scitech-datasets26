# Reflection

This dataset contains diagnostic measurements of cell nuclei taken from breast cancer biopsy samples, labeled as benign or malignant. It could be used to answer questions about which cellular features differ most between benign and malignant tumors. This dataset is relevant to genomic medicine because it supports disease classification and early detection based on biological measurements. One challenge was deciding which of the many available columns to focus on for analysis considering there were no null or n/a columns which would be my first choice to forego. Overall, the dataset was well structured and easy to explore using Pandas and Matplotlib.

Improved Graph Explanation: This bar chart illustrates the distribution of breast cancer diagnoses in the dataset, categorized as benign (B) and malignant (M). It presents the total number of samples within each diagnostic group. enumerate() was used to access both the index and the value of each diagnosis count so that numerical labels could be positioned accurately above each bar in the visualization. Examining class distribution is a critical step in medical data analysis which I've learned in my own practical applications in ob opportunnities, as significant imbalance between categories can affect statistical results and bias predictive models.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("brca.csv")

# Count diagnoses
diagnosis_counts = df["y"].value_counts()

# Create figure
plt.figure(figsize=(6,4))
bars = diagnosis_counts.plot(kind="bar")

plt.title("Distribution of Breast Cancer Diagnoses")
plt.xlabel("Diagnosis Type")
plt.ylabel("Number of Samples")

# Add labels above bars
for i, value in enumerate(diagnosis_counts):
    plt.text(i, value + 5, str(value), ha='center')

# Required basic cleaning
df["y"].isnull().sum()

plt.xticks(ticks=[0,1], labels=["Benign (B)", "Malignant (M)"], rotation=0)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("brca.csv")
total_samples = len(df)
benign = diagnosis_counts["B"]
malignant = diagnosis_counts["M"]

print("Total samples:", total_samples)
print("Benign percentage:", round((benign/total_samples)*100, 2), "%")
print("Malignant percentage:", round((malignant/total_samples)*100, 2), "%")
```
